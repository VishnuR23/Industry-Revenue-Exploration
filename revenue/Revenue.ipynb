```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("revenue.csv")
```

```python
df.head()
```

```python
df = df.drop(["State-owned", "Ref.", "Unnamed: 0"], axis="columns")
```

```python
df = pd.concat([df.drop('Headquarters[note 1]', axis = 1), pd.get_dummies(df['Headquarters[note 1]']).add_prefix('Headquarters_')], axis = 1)
```

```python
df
```

```python
df.columns.values
```

```python
df['Revenue'] = df['Revenue'].replace({'\$': '', ',':''}, regex=True)
df['Revenue'] = df['Revenue'].astype(int)
```

```python
df
```

```python
df['Profit'] = df['Profit'].replace({'\$':'', ',':'', '–': '-', '—': '-', '\u2212': '-'}, regex=True)
df['Profit'] = df['Profit'].astype(int)
```

```python
df
```

```python
df['Employees'] = df['Employees'].replace({',':''}, regex=True)
df['Employees'] = df['Employees'].astype(int)
```

```python
df['Revenue per worker'] = df['Revenue per worker'].replace({'\$': '', ',':''}, regex=True)
df['Revenue per worker'] = df['Revenue per worker'].astype(float)
df['Revenue per worker'] = df['Revenue per worker'].astype(int)
```

```python
df
```

```python
df = pd.concat([df.drop('Industry', axis = 1), pd.get_dummies(df['Industry']).add_prefix('Industry')], axis = 1)
```

```python
df
```

```python
df = df.drop('Name', axis=1)
```

```python
df.corr()
```

```python
correlations = df.corr()['Revenue'].abs()
sorted_corr = correlations.sort_values()
num_cols_to_drop = int(0.8 * len(df.columns))
cols_to_drop = sorted_corr.iloc[:num_cols_to_drop].index
df_dropped = df.drop(cols_to_drop, axis=1)
```

```python
df_dropped
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_dropped.corr(),annot=True,cmap='coolwarm')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
train_df,test_df = train_test_split(df,test_size=0.2)
```

```python
train_df
```

```python
test_df
```

```python
train_x = train_df.drop('Revenue', axis = 1)
train_y = train_df['Revenue']

test_x = test_df.drop('Revenue', axis = 1)
test_y = test_df['Revenue']
```

```python
forest = RandomForestClassifier()
forest.fit(train_x, train_y)
```

```python
forest.feature_importances_
```

```python
forest.feature_names_in_
```

```python
importances = dict(zip(forest.feature_names_in_, forest.feature_importances_))
importances = {k: v for k, v in sorted(importances.items(), key = lambda x: x[1], reverse=True)}
```
